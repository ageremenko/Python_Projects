# Корреляционный анализ

```python
def corr_analysis(df, target_name):    
    # создаём общую таблицу корреляции
    main_corr_matrix = df.corr()

    # задаём коэффициенты корреляции для таргетной переменной и для множественной корреляции между аттрибутами
    coef_for_target = 0.05
    coef_for_multicorr = 0.5

    # создаём таблицу корреляции, где корреляция с целевым признаком больше coef_for_target
    print("Работаем с целевым признаком "+str(target_name))
    corr_matrix = main_corr_matrix[abs(main_corr_matrix[target_name])>=coef_for_target]\
    [list(main_corr_matrix[abs(main_corr_matrix[target_name])>=coef_for_target].index)]

    # отбираем названия атрибутов, отсортированные по убыванию значений корреляции со статусом
    attr_by_corr_target = corr_matrix.sort_values(target_name, ascending=False).index[1:]

    # создаём таблицу кореляции, где и индекс, и колонки отсортированы по убыванию значений корреляции с таргетной переменной
    current_corr_matrix = corr_matrix.sort_values(target_name,ascending=False)[attr_by_corr_target][1:]

    flag = True

    # ходим по таблице и удаляем один из коррелирующих атрибутов до тех пор,
    # пока корреляция между всеми атрибутами не будет меньше coef_for_multicorr
    while flag:
        flag2 = False
        flag3 = False
        for i in current_corr_matrix.columns:
            for j in current_corr_matrix.columns:
                if (abs(current_corr_matrix[i][j])>=coef_for_multicorr) & (i!=j):
                    print('Сравниваем '+i+'='\
                          +str(corr_matrix[target_name][i])+' и '\
                          +j+'='+str(corr_matrix[target_name][j]))

                    if abs(corr_matrix[target_name][i]) < abs(corr_matrix[target_name][j]):
                        print('Удаляем '+i+'='+str(corr_matrix[target_name][i]))
                        current_corr_matrix.drop(i, axis=1, inplace=True)
                        current_corr_matrix.drop(i, axis=0, inplace=True)
                        flag2=True
                    else:
                        print('Удаляем '+j+'='+str(corr_matrix[target_name][j]))
                        current_corr_matrix.drop(j, axis=1, inplace=True)
                        current_corr_matrix.drop(j, axis=0, inplace=True)
                        flag2=True
                    print() 
                if flag2:
                    break
            if flag2:
                    break

        # ещё раз проходим по матрице для поиска атрибутов, корреляция которых больше coef_for_multicorr       
        for i in current_corr_matrix.columns:
            for j in current_corr_matrix.columns:
                if (abs(current_corr_matrix[i][j])>=coef_for_multicorr) & (i!=j):
                    flag3=True
                if flag3:
                    break
            if flag3:
                break 
        if flag3:
            continue

        flag = False
    
    fig, ax = plt.subplots(figsize=(10,10))
    sns.heatmap(abs(current_corr_matrix))

    return list(current_corr_matrix.index)
```

### Отбираем аттрибуты по корреляции с таргетными переменными больше 0.05 и корреляцией между собой меньше 0.5

```python
target_variables = ['B_FL_RM', 'B_FL_R05', 'B_FL_A5_TOT', 'B_FL_A5_UNF', 'LN_B_FL_RM', 'LN_B_FL_R05']

list_of_attributes = []

for i in target_variables:
    print("Оставшиеся аттрибуты"+str(corr_analysis(df, i)))
    print('')
    list_of_attributes = list_of_attributes + corr_analysis(df, i)
```

## При анализе графиков зависимости целевых признаков от параметров технологии явных зависимостей не было обнаружено

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy import stats
```

```python
for target in target_variables:
    data = df[target].dropna()  # Удаляем строки с пропущенными данными
    # Визуализируем распределение
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.hist(data, bins=20, color='skyblue', edgecolor='black')
    plt.title(f'{target} - Гистограмма')
    
    # QQ график для нормализации
    plt.subplot(1, 2, 2)
    sm.qqplot(data, line='s', color='skyblue')
    plt.title(f'{target} - QQ график')
    
    #  Тест Шапиро—Уилка на нормальность распределения
    shapiro_test = stats.shapiro(data)
    print(f'{target} - Критерий Шапиро—Уилка (W-statistic, p-value): {shapiro_test}')
    
    plt.show()
```

### Если учитывать, что размерность выборки составляет ~2000, то уровень значимости следует брать следующий: $\alpha$=0,001
### Так как для всех целевых признаков p-value < $\alpha$ , то это позволяет сделать вывод о статистической значимости данных свойств проката. Однако из всех представленных признаков только по гистограммам целевых признаков B_FL_RM, B_FL_A5_UNF и LN_B_FL_RM можно сделать вывод о приближенности данных к нормальному распределению.
