```python
data = pd.read_csv("Tesla.csv", parse_dates=True)
data.info()
data
```

```python
db = data.loc[(data['Date']>="2014-01-01")&(data['Date']<="2014-12-31")]
plt.figure(figsize = (24, 10))
plt.plot(db['Date'],db['High'], label='High')
plt.plot(db['Date'],db['Low'], label='Low')
buy_price = 250
plt.axhline(y=buy_price, color='r', ls='--', label='Buy price {}$'.format(buy_price))
plt.xticks(db['Date'].iloc[::3], rotation='vertical')
plt.yticks(range(int(db['Low'].min())//10*10, int(db['High'].max())//10*10+10, 5))
plt.grid()
plt.legend(loc='lower right', title='Legend')
```

```python
data2 = pd.read_csv("AB_NYC_2019.csv", parse_dates=True)
data2
```

```python
plt.figure(figsize=(20, 10))
for name, group in data2.groupby(['neighbourhood']):
    plt.scatter(group['longitude'], group['latitude'], s=0.1, label=name)
plt.xlabel('longitude')
plt.ylabel('latitude')    
plt.legend(markerscale=10)
plt.title('Accomodation location by neighbourhood')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(data2['longitude'], data2['latitude'], s=0.1, c=data2['availability_365'])
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar()
plt.title('Accomodation location by year availability')
plt.show()
```

```python

import numpy as np
import pandas as pd
import math
import missingno as msno
from datetime import datetime, timedelta

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Set2')

import os

df_kiva_loans = pd.read_csv("kiva_loans.csv")
df_mpi = pd.read_csv("kiva_mpi_region_locations.csv")
df_kiva_loans.head()
```

```python
df_kiva_loans_male = df_kiva_loans.loc[df_kiva_loans['borrower_genders']=='male']

regions_m = df_kiva_loans_male['world_region'].value_counts()

regions_list_m = regions_m.index.to_list()

dev_m = (df_kiva_loans_male['loan_amount']-df_kiva_loans_male['loan_amount'].mean()).abs()
# Стандартное отклонение
std_m = df_kiva_loans_male['loan_amount'].std()
# Фильтруем исходный набор данных
df_kiva_loans_male_trimmed = df_kiva_loans_male[~(dev_m>3*std_m)]

plt.figure(figsize=(16,10))
plt.title("Суммы займов по макрорегионам для мужчин", fontsize=18)
plt.xticks(np.arange(0, max(df_kiva_loans_male['loan_amount']+1), 500))
plt.tick_params(labelsize=14)

sns.boxplot(x='loan_amount', y="world_region", order=regions_list_m, data=df_kiva_loans_male_trimmed)\
    .set(ylabel=None, xlabel=None)

plt.show()

df_kiva_loans_female = df_kiva_loans.loc[df_kiva_loans['borrower_genders']=='female']

regions_f = df_kiva_loans_female['world_region'].value_counts()

regions_list_f = regions_f.index.to_list()

dev_f = (df_kiva_loans_female['loan_amount']-df_kiva_loans_female['loan_amount'].mean()).abs()
# Стандартное отклонение
std_f= df_kiva_loans_female['loan_amount'].std()
# Фильтруем исходный набор данных
df_kiva_loans_female_trimmed = df_kiva_loans_female[~(dev_f>3*std_f)]

plt.figure(figsize=(16,10))
plt.title("Суммы займов по макрорегионам для женщин", fontsize=18)
plt.xticks(np.arange(0, max(df_kiva_loans_female['loan_amount']+1), 500))
plt.tick_params(labelsize=14)

sns.boxplot(x='loan_amount', y="world_region", order=regions_list_f, data=df_kiva_loans_female_trimmed)\
    .set(ylabel=None, xlabel=None)

plt.show()

df_kiva_loans_group = df_kiva_loans.loc[df_kiva_loans['borrower_genders']=='group']

regions_g = df_kiva_loans_group['world_region'].value_counts()

regions_list_g = regions_g.index.to_list()

dev_g = (df_kiva_loans_group['loan_amount']-df_kiva_loans_group['loan_amount'].mean()).abs()
# Стандартное отклонение
std_g = df_kiva_loans_group['loan_amount'].std()
# Фильтруем исходный набор данных
df_kiva_loans_group_trimmed = df_kiva_loans_group[~(dev_g>3*std_g)]

plt.figure(figsize=(16,10))
plt.title("Суммы займов по макрорегионам для групп", fontsize=18)
plt.xticks(np.arange(0, max(df_kiva_loans_group['loan_amount']+1), 500))
plt.tick_params(labelsize=14)

sns.boxplot(x='loan_amount', y="world_region", order=regions_list_g, data=df_kiva_loans_group_trimmed)\
    .set(ylabel=None, xlabel=None)

plt.show()
```
