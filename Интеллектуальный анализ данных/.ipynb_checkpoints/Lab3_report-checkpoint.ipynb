```python
import pandas as pd
import sqlite3
```

#### 1. При помощи модуля sqlite3 откройте базу данных Instacart в файле instacart.db.

```python
conn = sqlite3.connect('instacart.db')
cursor = conn.cursor() 
```


#### 2. При помощи запроса SELECT извлеките из таблицы order_products__train записи, соответствующие указанным в индивидуальном задании дню недели (поле order_dow таблицы orders) и коду департамента (поле department_id таблицы products). Определите количество записей в полученном наборе и определите количество товаров (поле product_id таблицы order_products__train) в транзакциях набора.

```python
for row in conn.execute("""
    SELECT count(*),sum(department_id)
    FROM(SELECT *
    FROM orders as ord,order_products__train as ord_prod,products as prod
    WHERE ord_prod.product_id=prod.product_id AND ord_prod.order_id=ord.order_id AND
    ord.order_dow='5' AND prod.department_id='5')
    """):
    print(row)
```

#### 3. Определите количество покупок (транзакций) для пяти наиболее популярных товаров в наборе.

```python
for row in cursor.execute("""
    select count(opp.order_id) as orders, p.product_name as popular_product
    from order_products__train opp, products p
    where p.product_id = opp.product_id 
    group by popular_product 
    order by orders desc 
    limit 5
    """):
    print(row)
```

#### 4. Постройте транзакционную базу данных для поиск ассоциативных правил из полученного набора записей таблицы order_products__train, используя в качестве идентификатора транзакции поле order_id, а в качестве названий товаров - поле product_name из таблицы products, соответствующее полю product_id.

```python
for row in conn.execute("""
    SELECT ord_prod.order_id,prod.product_id,prod.product_name
    FROM orders as ord,order_products__train as ord_prod,products as prod
    WHERE ord_prod.product_id=prod.product_id AND ord_prod.order_id=ord.order_id AND
    ord.order_dow='5' AND prod.department_id='5'
    ORDER BY ord_prod.order_id
    """):
    print(row)
```
