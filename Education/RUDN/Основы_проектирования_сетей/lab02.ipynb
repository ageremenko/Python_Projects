```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = 30          # пиковая пропускная способность соты
g = 28          # полнодоступная часть пропускной способности соты
lambda_1 = 30    # интенсивность поступления запросов на предоставление услуги 1-го типа 
lambda_2 = 0   # интенсивность поступления запросов на предоставление услуги 2-го типа
mu = 0.5      # среднее время обслуживания запроса на предоставление услуги 1-го, 2-го типа
```

```python
def stat_dist (c, g, mu, lambda_1, lambda_2):   # стационарное распределение вероятностей состояний системы
    
    ro_1 = lambda_1/mu
    ro_2 = lambda_2/mu
    
    p = []
    p_0 = 0
    p_n = 0
    
    for i in range(0, g+1):
        p_0 += (pow(ro_1+ro_2, i))/math.factorial(i)
    
    for i in range(g+1, c+1):
        p_0 += (pow(ro_1+ro_2, g)*pow(ro_1, i-g))/math.factorial(i)
    
    p_0 = pow(p_0, -1)
        
    for n in range(0, g+1):
        p_n = (pow(ro_1+ro_2, n))/math.factorial(n)
        p.append(p_0*p_n)
        
    for n in range(g+1, c+1):
        p_n = (pow(ro_1+ro_2, g)*pow(ro_1, n-g))/math.factorial(n)
        p.append(p_0*p_n)
        
    return(p)
```

Стационарное распределение вероятностей состояний системы $p_n, n=1,...,50$:

```python
count = 0
for result in stat_dist(c, g, mu, lambda_1, lambda_2):
    print(result)
    count +=1
```

Проверка условия нормировки $\sum_{n=0}^C p_n =1$:

```python
print("sum(stat_dist(c, g, mu, lambda_1, lambda_2) = "\
      +str(sum(stat_dist(c, g, mu, lambda_1, lambda_2)))) # условие нормировки
```

```python
def time_blok_prob(c, g, mu, lambda_1, lambda_2):
    
    e = []
    p = stat_dist(c, g, mu, lambda_1, lambda_2)
    
    e.append(p[c])
    e.append(sum(p[g:c+1]))
    
    return e
```

```python
print("Вероятность блокировки по времени: "\
      +str(time_blok_prob(c, g, mu, lambda_1, lambda_2)))
```

```python
def call_blok_prob(c, g, mu, lambda_1, lambda_2):
    
    e = time_blok_prob(c, g, mu, lambda_1, lambda_2)
    b_1 = lambda_1*e[0]/(lambda_1+lambda_2)
    b_2 = lambda_2*e[1]/(lambda_1+lambda_2)
    
    return [b_1,b_2]
```

```python
print("Вероятности блокировки по вызовам: "\
      +str(call_blok_prob(c, g, mu, lambda_1, lambda_2)))
```

```python
def load_blok_prob(c, g, mu, lambda_1, lambda_2):
    
    C = time_blok_prob(c, g, mu, lambda_1, lambda_2)
    
    return C
    
```

```python
print("Вероятность блокировки по нагрузке: "\
      +str(load_blok_prob(c, g, mu, lambda_1, lambda_2))) 
```

```python
def avg_req(c, g, mu, lambda_1, lambda_2):
    
    n = 0
    p = stat_dist(c, g, mu, lambda_1, lambda_2)
    
    for i in range(c+1):
        n += i*p[i]
    
    return n
```

```python
print("Среднее число заявок: "\
      +str(avg_req(c, g, mu, lambda_1, lambda_2))) 
```

```python
load_prob = []
acc = 0.1
num = int(c/acc)


for i in np.linspace(lambda_1-1,lambda_2+1,num):
    load_prob.append(load_blok_prob(c, g, mu, lambda_1, i))
plt.figure(figsize=(6,3), dpi=100)
plt.plot(np.linspace(lambda_1-1,lambda_2+1, num), load_prob)


plt.xlabel('Интенсивность поступления запросов на обслуживание при $\lambda_1$ = '\
           +str(lambda_1)+' и '+'$\lambda_2$ = '+str(lambda_1)+',..,'+str(lambda_2))
plt.ylabel('Вероятность блокировки по нагрузке')
plt.show()
```
