```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = 50          # пиковая пропускная способность соты
lambda_1 = 1    # интенсивность поступления запросов на предоставление услуги 1-го типа 
lambda_2 = 20   # интенсивность поступления запросов на предоставление услуги 2-го типа
mu = 0.25      # среднее время обслуживания запроса на предоставление услуги 1-го, 2-го типа
```

```python
def stat_dist (c,  mu, lambda_1, lambda_2):   # стационарное распределение вероятностей состояний системы
    
    ro_1 = lambda_1/mu
    ro_2 = lambda_2/mu
    
    p = []
    p_a = 0
    p_b = 0
    
    for i in range(0, c+1):
        p_a += (pow(ro_1+ro_2, i))/math.factorial(i)
    
    p_a = pow(p_a, -1)
        
    for n in range(0, c+1):
        p_b = (pow(ro_1+ro_2, n))/math.factorial(n)
        p.append(p_a*p_b)
        
    return(p)
```

Стационарное распределение вероятностей состояний системы $p_n, n=1,...,50$:

```python
for result in stat_dist(c,  mu, lambda_1, lambda_2):
    print(result)
```

Проверка условия нормировки $\sum_{n=0}^C p_n =1$:

```python
print("sum(stat_dist(c,  mu, lambda_1, lambda_2) = "\
      +str(sum(stat_dist(c,  mu, lambda_1, lambda_2)))) # условие нормировки
```

```python
def time_blok_prob(c,  mu, lambda_1, lambda_2):
    
    p = stat_dist(c,  mu, lambda_1, lambda_2)
    e = p[c]
    
    return e
```

```python
print("Вероятность блокировки по времени: "\
      +str(time_blok_prob(c,  mu, lambda_1, lambda_2)))
```

```python
def call_blok_prob(c,  mu, lambda_1, lambda_2):
    
    e = time_blok_prob(c,  mu, lambda_1, lambda_2)
    b_1 = lambda_1*e/(lambda_1+lambda_2)
    b_2 = lambda_2*e/(lambda_1+lambda_2)
    
    return [b_1,b_2]
```

```python
print("Вероятности блокировки по вызовам: "\
      +str(call_blok_prob(c,  mu, lambda_1, lambda_2)))
```

```python
def load_blok_prob(c,  mu, lambda_1, lambda_2):
    
    p = stat_dist(c,  mu, lambda_1, lambda_2)
    c = time_blok_prob(c,  mu, lambda_1, lambda_2)
    
    return c
    
```

```python
print("Вероятность блокировки по нагрузке: "\
      +str(load_blok_prob(c,  mu, lambda_1, lambda_2))) 
```

```python
def avg_req(c,  mu, lambda_1, lambda_2):
    
    n = 0
    p = stat_dist(c,  mu, lambda_1, lambda_2)
    
    for i in range(c+1):
        n += i*p[i]
    
    return n
```

```python
print("Среднее число заявок: "\
      +str(avg_req(c,  mu, lambda_1, lambda_2))) 
```

```python
time_prob = []
acc = 0.1
num = int(c/acc)

for i in np.linspace(lambda_1-1,lambda_2+1,num):
    time_prob.append(call_blok_prob(c, mu, lambda_1, i))

plt.figure(figsize=(6,3), dpi=100)
plt.plot(np.linspace(lambda_1-1,lambda_2+1, num), time_prob)
plt.xlabel('Интенсивность поступления запросов на обслуживание при $\lambda_1$ = '\
           +str(lambda_1)+' и '+'$\lambda_2$ = '+str(lambda_1)+',..,'+str(lambda_2))
plt.ylabel('Вероятность блокировки по времени')
plt.show()
```

```python
time_prob = []

for i in np.linspace(lambda_1-1,lambda_2+1,num):
    time_prob.append(call_blok_prob(c, mu, i, lambda_2))

plt.figure(figsize=(6,3), dpi=100)
plt.plot(np.linspace(lambda_1-1,lambda_2+1, num), time_prob)
plt.xlabel('Интенсивность поступления запросов на обслуживание при $\lambda_1$ = '\
           +str(lambda_1)+',..,'+str(lambda_2)+' и '+'$\lambda_2$ = '+str(lambda_2))
plt.ylabel('Вероятность блокировки по времени')
plt.show()
```

```python
avg_re = []

for j in np.linspace(lambda_1-1,lambda_2+1,num):
    avg_re.append(avg_req(c, mu, lambda_1, j))

plt.figure(figsize=(6,3), dpi=100)
plt.plot(np.linspace(lambda_1-1,lambda_2+1, num), avg_re)
plt.xlabel('Интенсивность поступления запросов на обслуживание при $\lambda_1$ = '+str(lambda_1)+' и '+'$\lambda_2$ = '\
           +str(lambda_1)+',..,'+str(lambda_2))
plt.ylabel('Среднее число обслуживаемых запросов')
plt.show()
```
