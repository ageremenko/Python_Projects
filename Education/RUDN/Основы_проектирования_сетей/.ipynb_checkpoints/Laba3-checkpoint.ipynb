```python
def fctr(i):
    fc = 1
    for j in range(2,i+1):
        fc = fc*j;
    return fc
```

```python
import numpy as np
import matplotlib.pyplot as plt

C = 40 

lya_min = 1 #Начальное Лямбда 1
lya_max = 40 #Конечное Лямбда 1
lya1 = [] #массив для Lambda_1
lya2 = 30
o1= 2.5 #ТЭТА
o2 = 2


klv1 = C
klv2 = C


a2 = lya2*o2/C
a1 = []
NN1 = []
T1 = []
NN2 = []
T2 = []
veroy = [[]]

veroy= np.zeros((klv1+1, klv2+1), dtype=float).tolist()

lya1.append(lya_min)

shag = 0.2 #шаг для Lambda_1

while(lya_min < lya_max):
    lya_min = round(lya_min+shag,1)
    lya1.append(lya_min)
  
    
for i in range(0, (len(lya1))):
    a1.append(lya1[i]*o1/C)
    NN1.append(0)
    NN2.append(0)
    T1.append(0)
    T2.append(0)
```

```python
for i in range(0, len(lya1)):
    
    veroy[0][0]=0
    for n1 in range(0, klv1+1):
        for n2 in range(0, klv2+1):
            veroy[0][0] = veroy[0][0] + fctr(n1+n2)*(a1[i]**n1)*(a2**n2)/(fctr(n1)*fctr(n2))
               
    veroy[0][0]=1/veroy[0][0]

    for n1 in range(0, klv1+1):
        for n2 in range(0, klv2+1):
            if (n1==0 and n2==0):
                continue
            else:
                veroy[n1][n2]=veroy[0][0]*fctr(n1+n2)*(a1[i]**n1)*(a2**n2)/(fctr(n1)*fctr(n2))
               
    NN1[i]=lya1[i]*o1/(lya1[i]*o1+lya2*o2)
    NN2[i]=lya2*o2/(lya1[i]*o1+lya2*o2)
    
    T1[i]=NN1[i]/lya1[i]
    T2[i]=NN2[i]/lya2

print("Распределение вероятностей для C=%d, lya1=%d, lya2=%d, o1=%s, o2=%s:" %(C, lya1[len(lya1)-1], lya2, o1, o2))
for n1 in range(0, klv1+1):
        for n2 in range(0, klv2+1):
                print('veroy[%d][%d]=%s'%(n1,n2,veroy[n1][n2]))
              
print("Среднее число запросов:",NN1[len(lya1)-1], NN2[len(lya1)-1])
print("Среднее время обслуживания:",T1[len(lya1)-1], T2[len(lya1)-1])
```

```python
fig = plt.figure(figsize=(10,10))
plt.plot(lya1, NN1, label = "на передачу данных первого типа")
plt.plot(lya1, NN2, label = "на передачу данных второго типа")
plt.title("Среднее число обслуживаемых в системе запросов")
plt.xlabel("Lambda_1")
plt.ylabel("Amount")
plt.legend()
```

```python
fig = plt.figure(figsize=(10,10))
plt.plot(lya1, T1, label = "на передачу данных первого типа")
plt.plot(lya1, T2, label = "на передачу данных второго типа")
plt.title("Среднее время обслуживания запроса")
plt.xlabel("Lambda_1")
plt.ylabel("Time")
plt.legend()
```
