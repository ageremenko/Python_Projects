```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import LinearSVC
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

def PolynomialLogisticRegression(degree):
    return Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('std_scaler', StandardScaler()),
        ('log_reg', LogisticRegression())
    ])

def PolynomialLinearSVC(degree):
    return Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('std_scaler', StandardScaler()),
        ('svc', LinearSVC())
    ])

def plot_decision_boundary(model, axis):
    
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1, 1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1, 1),
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
    
    plt.contourf(x0, x1, zz, cmap=custom_cmap)
```

3.         Вычислите и визуализируйте матрицу корреляций признаков. Удалите из набора признаки, имеющие высокую корреляцию (близкую к +1 или -1) с другими признаками.

 



```python
import warnings
warnings.filterwarnings("ignore")

def correlation_matrix(df):
    from matplotlib import cm as cm

    fig = plt.figure(figsize=(16,12))
    ax1 = fig.add_subplot(111)
    cmap = cm.get_cmap('jet', 30)
    cax = ax1.imshow(df.corr(), cmap=cmap)
    ax1.grid(True)
    plt.title('Корреляции признаков в наборе данных движений языка жестов LIBRAS\n',fontsize=15)
    #labels=df.columns
    #ax1.set_xticklabels(labels,fontsize=9)
    #ax1.set_yticklabels(labels,fontsize=9)
    fig.colorbar(cax, ticks=[0.1*i for i in range(-11,11)])
    plt.show()

correlation_matrix(main_df.drop(90,axis=1))
```

```python
corr_matrix = main_df.drop(90,axis=1).corr()         # корреляционная матрица
corr_list = list(main_df.drop(90,axis=1).columns)   # массив с индексами всех столбцов 
                                                    
for i in corr_matrix.columns:
    for j in corr_matrix.columns:         
        if (abs(corr_matrix[i][j])>0.9826)&(corr_matrix[i].name in corr_list)&(len(corr_list)>0)&(i!=j): 
            corr_list.remove(i)     # удаляем индексы столбцов, у которых значения корреляции выше чем подобранное вручную

df = pd.DataFrame(columns=corr_list, data = main_df) # берём только отобранные столбцы из датафрейма
df
```

```python
classes = main_df[90]

list_class_0 = [3, 4, 5, 6, 15]

for i in range(0,len(classes)):
    if classes[i] in list_class_0:
        classes[i] = 0
    else:
        classes[i] = 1

classes.name = 'class'
classes
```

5.         Используя метод рекурсивного исключения признаков (RFE) и логистическую регрессию, определите и оставьте в наборе наиболее значимые признаки (не менее двух). Если в наборе данных осталось более двух признаков, то определите два признака с наибольшей дисперсией для визуализации.

 



```python
#RFE
X = df
y = classes
# feature extraction
model = LogisticRegression(max_iter=1000)
rfe = RFE(model)
fit = rfe.fit(X, y);
```

```python
print("Число признаков: %d" % fit.n_features_)
print("Выбранные признаки: %s" % fit.support_)
print("Ранг признаков: %s" % fit.ranking_)
```

```python
RFE_df = df[df.columns[fit.ranking_==1]]
RFE_df
```

```python
X = RFE_df
y = classes
target_names = ['Class 1', 'Class 2']
y
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit(X).transform(X)
```

```python
plt.figure(figsize=(10,8))

for  i, target_name in zip([0, 1], target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], alpha=.8, lw=2,
                label=target_names[i])
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('Метод PCA для набора Libras Movement');
```

6.         Масштабируйте признаки набора данных на интервал от 0 до 1.

 



```python
X_pca[:,0] = (X_pca[:,0] - np.min(X_pca[:,0])) / (np.max(X_pca[:,0]) - np.min(X_pca[:,0]))
X_pca[:,1] = (X_pca[:,1] - np.min(X_pca[:,1])) / (np.max(X_pca[:,1]) - np.min(X_pca[:,1]))
```

9.         Визуализируйте на одном рисунке ROC кривые для каждого из классификаторов, подписывая оси и рисунок и создавая легенду для методов бинарной классификации.

 



```python
# наивный байессовский классификатор почему-то не имеет decision_function, поэтому я его исключил

classifiers = [log_reg, poly_log_reg_2, poly_log_reg_3, lda, svc, poly_svc_2, poly_svc_3]
classifiers_names = ['Logistic Regression', 'Polynomial Logistic Regression, degree=2',\
                     'Polynomial Logistic Regression, degree=3', 'LDA', 'SVC',\
                     'Polynomial SVC, degree=2', 'Polynomial SVC, degree=3']
roc_auc_score_list = []

plt.figure(figsize=(10,8))

for i in range(len(classifiers)):

    decision_scores = classifiers[i].decision_function(X_test)
    fprs, tprs, thresholds = roc_curve(y_test, decision_scores)
    
    roc_auc_score_list.append(roc_auc_score(y_test, decision_scores))
    
    plt.plot(fprs, tprs, label=classifiers_names[i])
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    
plt.xlabel('FPR', fontsize=18)
plt.ylabel('TPR', fontsize=18)
plt.title('ROC-кривые')
plt.show()
```

10.       Определите лучший метод бинарной классификации набора данных по показателю ROC_AUC (площади под ROC кривой).

```python
final_df = pd.DataFrame(columns=['classifiers', 'classifiers_names', 'roc_auc_score'])
final_df['classifiers'] = classifiers
final_df['classifiers_names'] = classifiers_names
final_df['roc_auc_score'] = roc_auc_score_list
final_df
```

```python
print('Лучший метод бинарной классификации: '+\
      str(final_df[final_df['roc_auc_score']==final_df['roc_auc_score'].max()]['classifiers_names'].values[0]))
```
