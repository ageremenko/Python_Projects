```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv("ap-northeast-1.csv", header=None, parse_dates=True)
data.columns = ['date','arch','OS','zone', 'price']
data
```

```python
dt = data.loc[(data['OS']=='Windows') & ((data['arch']>='i') & (data['arch']<'j'))]
dt = dt.sort_values("date", ascending = True)
dt
```

```python
import math

i = 0
n = dt['arch'].nunique()//2
ar = pd.array(dt['arch'].unique())
fig, axes = plt.subplots(n, 2, figsize=(20,20))
plt.subplots_adjust(hspace=0.2)
for row in range(n):
    for column in range(2):
        arch = ar[i]
        dates = dt['date'].loc[dt['arch']==arch]
        prices = dt['price'].loc[dt['arch']==arch]
        
        axes[row, column].plot(dates, prices, label=arch, color='r')
        
        
        axes[row, column].set_xlabel('Dates')
        axes[row, column].set_ylabel('Prices')
        
        
        axes[row, column].set_xticks(dates.iloc[::(dates.count()//6)])
        axes[row, column].set_xticklabels(dates.iloc[::(dates.count()//6)], rotation=8)
        
        yticks = np.arange(math.floor(min(prices)), math.ceil(max(prices)), 5)
        axes[row, column].set_yticks(yticks)
        axes[row, column].set_yticklabels(yticks)
        
        
        axes[row, column].grid()
        axes[row, column].legend(loc='upper left', title='Prices')
        
        i+=1
        
plt.show() 
```
