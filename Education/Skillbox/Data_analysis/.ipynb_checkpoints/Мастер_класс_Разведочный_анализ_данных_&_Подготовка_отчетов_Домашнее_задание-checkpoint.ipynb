## 4.1. Самостоятельная работа
***

- А в каких странах преобладают мужчины? Группа людей? (п. 2.2.2.)
- Изучите несколько займов с очень большими суммами. Что можно сказать о назначении этих заявок? В каких странах они оформлены? Каков гендерный состав заявителей? (п. 2.4.2.)

```python
countries = df_kiva_loans['country'].value_counts()
list_countries = list(countries.index)
df_gender_by_country_male = \
    df_kiva_loans[df_kiva_loans['country'].isin(countries.index.values)]\
    .groupby(['country', 'world_region', 'borrower_genders'])['borrower_genders']\
    .count()\
    .groupby(level=0).apply(lambda x: 100 * x / x.sum())\
    .unstack('borrower_genders')\
    .fillna(0)\
    .sort_values(by=['male', 'female', 'group'])



df_gender_by_country_male = df_gender_by_country_male[['male', 'female', 'group']]
df_gender_by_country_male = df_gender_by_country_male.loc[((df_gender_by_country_male['male']>
                                                            df_gender_by_country_male['female'])&
                                                           (df_gender_by_country_male['male']>
                                                            df_gender_by_country_male['group']))]


plot = df_gender_by_country_male.plot.barh(
        figsize=(20,10)
        , fontsize=16
        , stacked=True
        , title='Гендерная структура заемщиков, где мужчин больше остальных')
plot.title.set_size(18)
plot.legend(loc=1, bbox_to_anchor=(1.2, 1), fontsize=16)
plot.set(ylabel=None, xlabel=None)
plt.show()
```

### Выше, на диаграмме, преобладают cтраны с Ближнего Востока, Африки и Южной Америки. Этот факт свидетельствует о том, что в списке страны приемущественно с традиционным укладом жизни, а значит и патриархатом в семье. Следовательно финансами и организацией бизнеса занимается, в основном, мужская часть населения.

```python
df_gender_by_country_group = \
    df_kiva_loans[df_kiva_loans['country'].isin(countries.index.values)]\
    .groupby(['country','world_region', 'borrower_genders'])['borrower_genders']\
    .count()\
    .groupby(level=0).apply(lambda x: 100 * x / x.sum())\
    .unstack('borrower_genders')\
    .fillna(0)\
    .sort_values(by=['group','male', 'female'])

df_gender_by_country_group = df_gender_by_country_group[['group','male', 'female']]
df_gender_by_country_group = df_gender_by_country_group.loc[((df_gender_by_country_group['group']>
                                                              df_gender_by_country_group['male'])&
                                                             (df_gender_by_country_group['group']>
                                                              df_gender_by_country_group['female']))]

plot2 = df_gender_by_country_group.plot.barh(
        figsize=(20,10)
        , fontsize=16
        , stacked=True
        , title='Гендерная структура заемщиков, где групп больше остальных')
plot2.title.set_size(18)
plot2.legend(loc=1, bbox_to_anchor=(1.2, 1), fontsize=16)
plot2.set(ylabel=None, xlabel=None)
plt.show()
```

```python
genders = ['female', 'male', 'group']

for i in range(3):
    df_kiva_loans_x = df_kiva_loans.loc[df_kiva_loans['borrower_genders']==genders[i]]

    countries_x = df_kiva_loans_x['country'].value_counts()
    countries_list_x = countries_x.index.to_list()    
    
    dev_x = (df_kiva_loans_x['loan_amount']-df_kiva_loans_x['loan_amount'].mean()).abs()
    # Стандартное отклонение
    std_x = df_kiva_loans_x['loan_amount'].std()
    # Фильтруем исходный набор данных
    df_kiva_loans_trimmed_x = df_kiva_loans_x[~(dev_x>3*std_x)]
    
    
    #ax[i].tick_params(labelsize=13) 
    #sns.boxplot(x='loan_amount', y="country", order=countries_list_x, data=df_kiva_loans_trimmed_x, ax=ax[i])
    
    sns.catplot(x="loan_amount", y="country", col='borrower_genders', order=countries_list_x,
                data=df_kiva_loans_trimmed_x, kind="box", height=11, aspect=1)
    plt.xlabel("Amount of loan", size=15)
    plt.ylabel("Country", size=15)
    plt.title(genders[i], size=20)
    
plt.tight_layout()
plt.show()
```

### В диаграммах по странам сложно увидеть какую-то закономерность. Единственное, что можно сказать, так это максимальные суммы займов для женщин, мужчин и групп людей - примерно 3000, 5000 и 7000 $ соответственно.

```python

for i in range(3):
    df_kiva_loans_x = df_kiva_loans.loc[df_kiva_loans['borrower_genders']==genders[i]]

    regions_x = df_kiva_loans_x['world_region'].value_counts()  
    regions_list_x = regions_x.index.to_list()

    dev_x = (df_kiva_loans_x['loan_amount']-df_kiva_loans_x['loan_amount'].mean()).abs()
    # Стандартное отклонение
    std_x = df_kiva_loans_x['loan_amount'].std()
    # Фильтруем исходный набор данных
    df_kiva_loans_trimmed_x = df_kiva_loans_x[~(dev_x>3*std_x)]
    
    sns.catplot(x="loan_amount", y="world_region",col='borrower_genders', order=regions_list_x,
                data=df_kiva_loans_trimmed_x, kind="box", height=4, aspect=3)
    plt.xlabel("Amount of loan", size=15)
    plt.ylabel("World region", size=15)
    plt.title(genders[i], size=20)

    print(regions_x)
    
plt.tight_layout()
plt.show()

```

## Зависимость пола заёмщика от региона планеты, сектора экономики

```python
# Наблюдаемое отклонение от среднего
dev = (df_kiva_loans['loan_amount']-df_kiva_loans['loan_amount'].mean()).abs()
# Стандартное отклонение
std = df_kiva_loans['loan_amount'].std()
# Фильтруем исходный набор данных
df_kiva_loans_trimmed = df_kiva_loans[~(dev>3*std)]

df_loans_dates = df_kiva_loans_trimmed.dropna(subset=['disbursed_time', 'funded_time'],
                                              how='any', inplace=False)


dates = ['posted_time','disbursed_time']


df_loans_dates[dates] = df_loans_dates[dates].applymap(lambda x : x.split('+')[0])
df_loans_dates[dates] = df_loans_dates[dates].apply(pd.to_datetime)


df_loans_dates.loc[:, 'time_funding'] = (df_loans_dates['disbursed_time']-df_loans_dates['posted_time'])
df_loans_dates.loc[:, 'time_funding'] = df_loans_dates['time_funding'] / timedelta(days=1)
#df_loans_dates.loc[:, 'time_funding'] = df_loans_dates.loc[:, 'time_funding'].reset_index()

dev = (df_loans_dates['time_funding']-df_loans_dates['time_funding'].mean()).abs()
std = df_loans_dates['time_funding'].std()

df_loans_dates_trimmed = df_loans_dates[~(dev>3*std) & (df_loans_dates.loc[:, 'time_funding']>0)]
#df_loans_dates_trimmed = df_loans_dates_trimmed.reset_index()

df_country_median = df_loans_dates_trimmed.groupby(['world_region', 'country'], as_index=False)\
    .agg({'lender_count' : 'median','loan_amount' : 'median', 'time_funding' : 'median', 'term_in_months' : 'median'})\
    .reset_index()

df_country_median['monthly_repayment'] = df_country_median['loan_amount'] / df_country_median['term_in_months']
df_country_median['monthly_repayment_for_person'] = df_country_median['monthly_repayment']/df_country_median['lender_count']
df_country_median = df_country_median[df_country_median.country.isin(list_countries)].sort_values(by='lender_count')

df_country_median = df_country_median.sort_values(by='term_in_months')
#df_country_median = df_country_median.reset_index()

```

```python
f,ax=plt.subplots(1, 3, sharey=True, figsize=(35,20))

sns.barplot(y='country',x='lender_count', data=df_country_median, alpha=0.6, ax=ax[0])
ax[0].set_title('Медиана количества кредиторов', fontsize=30)
ax[0].set_xlabel('Количество кредиторов', fontsize=25)
ax[0].set_ylabel(None)
ax[0].tick_params(labelsize=23)

sns.barplot(y='country',x='loan_amount', data=df_country_median, alpha=0.6, ax=ax[1])
ax[1].set_title('Медиана суммы займов', fontsize=30)
ax[1].set_xlabel('$', fontsize=25)
ax[1].set_ylabel(None)
ax[1].tick_params(labelsize=23)

sns.barplot(y='country',x='term_in_months', data=df_country_median, alpha=0.6, ax=ax[2])
ax[2].set_title('Медиана сроков займов', fontsize=30)
ax[2].set_xlabel('Количество месяцев', fontsize=25)
ax[2].set_ylabel(None)
ax[2].tick_params(labelsize=23)

plt.tight_layout()
plt.show()
```
